## EDA

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Customer Analysis
# Distribution of customers by region
plt.figure(figsize=(12, 6))
region_counts = customers['Region'].value_counts().head(20)  # Top 20 regions
sns.barplot(x=region_counts.index, y=region_counts.values, palette='viridis')
plt.title("Number of Customers by Region (Top 20)")
plt.xlabel("Region")
plt.ylabel("Number of Customers")
plt.xticks(rotation=45, ha='right')
plt.tight_layout() 
plt.show()
```

```python
# Correcting the date format and parse the SignupDate column
customers['SignupDate'] = pd.to_datetime(customers['SignupDate'], format='%d-%m-%Y', errors='coerce')
customers['SignupYearMonth'] = customers['SignupDate'].dt.to_period('M')  # Year-Month format
signup_counts = customers.groupby(pd.Grouper(key='SignupDate', freq='4M')).size()
# Plotting the number of signups
plt.figure(figsize=(12, 6))
sns.lineplot(x=signup_counts.index, y=signup_counts.values, marker='o', color="orange")
plt.title("Number of Customer Signups (For Every 4 Months)", fontsize=16)
plt.xlabel("Year-Month", fontsize=14)
plt.ylabel("Number of Signups", fontsize=14)
# Formatting x-axis ticks for clarity
plt.xticks(signup_counts.index, labels=[date.strftime('%Y-%m') for date in signup_counts.index], rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Product Analysis
# Products by category
category_counts = products['Category'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=category_counts.index[:10], y=category_counts.values[:10], palette='plasma')
plt.title("Top 10 Product Categories")
plt.xlabel("Category")
plt.ylabel("Number of Products")
plt.xticks(rotation=45)
plt.show()
```

```python
# Top 10 most expensive products
products['Price'] = pd.to_numeric(products['Price'], errors='coerce')
top_expensive = products.nlargest(10, 'Price')
plt.figure(figsize=(10, 6))
sns.barplot(x=top_expensive['ProductName'], y=top_expensive['Price'], palette='mako')
plt.title("Top 10 Most Expensive Products")
plt.xlabel("Product Name")
plt.ylabel("Price (USD)")
plt.xticks(rotation=45)
plt.show()
```

```python
# Transaction Analysis
transactions['TransactionDate'] = pd.to_datetime(transactions['TransactionDate'], format='%d-%m-%Y %H:%M')
transactions['TransactionMonth'] = transactions['TransactionDate'].dt.to_period('M')
monthly_sales = transactions.groupby('TransactionMonth')['TotalValue'].sum()
```

```python
# Plot sales trends
plt.figure(figsize=(12, 6))
monthly_sales.plot(kind='line', marker='o', color='teal')
plt.title("Monthly Sales Trends")
plt.xlabel("Month")
plt.ylabel("Total Sales Value (USD)")
plt.grid(True)
plt.show()
```

```python
# Top 10 products by sales
top_products = transactions.groupby('ProductID')['TotalValue'].sum().nlargest(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_products.index, y=top_products.values, palette='cubehelix')
plt.title("Top 10 Products by Total Sales Value")
plt.xlabel("Product ID")
plt.ylabel("Total Sales Value (USD)")
plt.show()
```
